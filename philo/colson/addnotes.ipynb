```python
import re

basename = 'tlg0018.tlg002'
f = open(basename + '-notes.txt')
notelist = {}
curnote = ''
for l in f:
    l = re.sub('\s+$','',l)
    l = re.sub('<sup>([0-9]+)</sup>','-\g<1>',l)
    m = re.search('<note n="([^"]+)"',l)
    if(m):
        curnote = m[1]
        notelist[curnote] = l 
        continue
    if(l == ''):
        continue
    if(not curnote):
        continue
    notelist[curnote] =  notelist[curnote] + ' ' + l
    

f.close()

f = open(basename + '.engX.xml')
outf = open(basename + '.colson_eng.xml','w')
curpage = ''
for l in f:
    l = re.sub('\s+$','',l)
    m = re.search('<pb n="([0-9]+)"',l)
    if(m):
        curpage = m[1]
    if(not curpage):
        print(l,file=outf)
        continue
    l = re.sub('(target="[0-9]+)-([0-9]+)-([a-z0-9]+")','\g<1>.\g<2>.\g<3>',l)
    while(re.search('<ref target="([^"]+)"',l)):
        
        
        m = re.search('<ref target="([^"]+)"/>',l)
        curref = m[1]
        if(not curref in notelist):
            print('missing',curref)
            l = re.sub('<ref (target="[^"]+)"','<ref-\g<1>',l,1)
            continue
        l = re.sub('<ref target="([^"]+)"/>',notelist[curref],l,1)
    l = re.sub('(<note)','\n\g<1>',l)
    l = re.sub('</note>','</note>\n',l)
    print(l,file=outf)


f.close()
outf.close()

f.close()

f = open(basename + '.grcX.xml')
outf = open(basename + '.colson_grc.xml','w')
curpage = ''
for l in f:
    l = re.sub('\s+$','',l)
    l = re.sub('(target="[0-9]+)-([0-9]+)-([a-z0-9]+")','\g<1>.\g<2>.\g<3>',l)
    m = re.search('(<pb n="[0-9]+\.)([0-9]+)"',l)
    if(m):
        curpage = str(int(m[2]) )
   #     l = re.sub('(<pb n="[0-9]+\.)([0-9]+)"','\g<1>'+curpage+'"',l)
        
    if(not curpage):
        print(l,file=outf)
        continue
    l = re.sub('(target="[0-9])-([0-9]+)-([a-z0-9]+")','\g<1>.\g<2>.\g<3>',l)
    #m = re.search('(target="[0-9])-([0-9]+)-([a-z0-9]+")',l)
    #if(m):
    #    curpage = str(int(m[2])-1)
    #    l = re.sub('(target="[0-9])-([0-9]+)-([a-z0-9]+)','\g<1>.'+curpage+'.\g<3>',l)
    #print(l)
    while(re.search('<ref target="([^"]+)"',l)):
        
        
        m = re.search('<ref target="([^"]+)"/>',l)
        if(m):
            curref = m[1]
        if(not curref in notelist):
            print('missing',curref)
            l = re.sub('<ref (target="[^"]+)"','<ref-\g<1>"',l,1)
            continue
        l = re.sub('<ref target="([^"]+)"/>',notelist[curref],l,1)
    l = re.sub('(<note)','\n\g<1>',l)
    l = re.sub('</note>','</note>\n',l)
    l = re.sub('<ref-','<ref ',l)
    l = re.sub('</div><div','</div>\n\n<div',l)
    #l = re.sub('<ref-','<ref ',l)
    #l = re.sub('(.)(<l)','\g<1>\n\g<2>',l)
    print(l,file=outf)


f.close()
outf.close()
```
