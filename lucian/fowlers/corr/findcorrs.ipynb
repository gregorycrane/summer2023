```python
import re

pageerrs = {}

def additem(curitem,curdict):
    if(not curitem in curdict):
        curdict[curitem] = 1
    else:
        curdict[curitem] = curdict[curitem] + 1

def dofile(curvol):
    fname = 'corr-lucian-vol'+curvol+'-eng.xml'
    outfname = 'corrs-' + curvol + '.tsv'
    outf = open(outfname,'w')
    f = open(fname)
    global pageerrs
    errs = 0
    curpage = ''
    curpat = '([^ ]+\s+[^ ]+)\s+<choice><sic>([^<]+)</sic><corr>([^<]+)</corr></choice>\s+([^ ]+\s+[^ ]+)'
    for l in f:
        m = re.search('<pb n="([^"]+)"',l)
        if(m):
            curpage = m[1]
            
        while(re.search(curpat,l)):
            m = re.search(curpat,l)
            errs = errs + 1
            org = m[2]
            fix = m[3]
            errtype = 'UNK'
            if(re.sub(',','.',org) == fix):
                errtype = 'period'
            if(re.sub('-','',fix) == re.sub('-','',org)):
                errtype = 'hyphen'
            if(re.sub('\\bb(ad|e|is|im|it)\\b','h\g<1>',org) == fix):
                errtype = 'h->b'
            if(re.sub('\s+','',org) == re.sub('\s+','',fix)):
                errtype = 'space'
            print(errs,errtype,curpage,m[1],m[2],m[3],m[4],sep='\t',file=outf)
            l = re.sub(curpat,'',l,1)
            additem(curvol+'.'+curpage,pageerrs)
            
            

            
    
    
    f.close()
    outf.close()
    

for i in range(4,5):
    curvol = str(i)
    dofile(curvol)

for foo in sorted(pageerrs,key=pageerrs.get,reverse=True):
    print(foo,pageerrs[foo])
```
