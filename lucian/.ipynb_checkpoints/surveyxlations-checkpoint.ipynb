```python
import re

xlations = {}
xllens = {}
filelist = ['fowlers/corr/corr-lucian-vol1-eng.xml',
'fowlers/corr/corr-lucian-vol2-eng.xml',
'fowlers/corr/corr-lucian-vol3-eng.xml',
'fowlers/corr/corr-lucian-vol4-eng.xml',
'harmon/work2-lucian-loeb01.1st1k-eng1.xml',
'harmon/work2-lucian-loeb02.1st1k-eng1.xml',
'harmon/work2-lucian-loeb03.1st1k-eng1.xml',
'harmon/work2-lucian-loeb04.1st1k-eng1.xml',
'harmon/work2-lucian-loeb05.1st1k-eng1.xml',
'williams/lucian_1888-3.xml','deasyria/lucian-dedea.xml','hetaer/hetaer-eng.xml']

curauthor = 'tlg0062'
tottokens = 0
def getworks(fname):
    global xlations
    global tottokens
    f = open(fname)
    fullid = ''
    for l in f:
        if(re.search('<idno',l)):
            continue
        m = re.search('('+curauthor+'\.tlg0[0-9][0-9])\\b',l)
        if(m):
            curwork = m[1]
            print(curwork, fname,l)
            if(not curwork in xlations):
                xlations[curwork] = []
            xlations[curwork].append(fname)
            xlator = re.sub('^([a-z]+).+','\g<1>',fname)
            fullid = curwork + '.' + xlator
            #print(fullid)
        if(fullid):
            if(not fullid in xllens):
                xllens[fullid] = 0
            xllens[fullid] = len(l.split()) + xllens[fullid] 
            tottokens = tottokens + len(l.split())
    
    f.close()
    
for foo in filelist:
    getworks(foo)
```

```python
for i in range(1,72):
    if(i<10):
        curs = curauthor + '.tlg00' + str(i)
    else:
        curs = curauthor + '.tlg0' + str(i)

    if(not curs in xlations):
        print(curs,0)
    else:
        print(curs,len(xlations[curs]),xlations[curs])
```
