```python
import re

firstpages = {}

def regacc(s):
    s = re.sub('δι᾿',"δι’",s)
    s = re.sub('ά','ά',s)
    s = re.sub('έ','έ',s)
    s = re.sub('ή','ή',s)
    s = re.sub('ό','ό',s)
    s = re.sub('ώ','ώ',s)
    s = s.replace('ί','ί')
    s = s.replace('ί','ί')
    s = s.replace('ώ','ώ')
    s = s.replace('ύ','ύ')
    s = s.replace('ύ','ύ')
    
    s = s.replace('·','·')
    s = s.replace('ἀ','ἀ')
    s = s.replace(',,','„')
    s = s.replace('ἐ','ἐ')
    s = s.replace('’Α','Ἀ')
    s = s.replace('᾿Ε','Ἐ')
    s = s.replace('᾿Η','Ἠ')
    s = s.replace('᾿Ι','Ἰ')
    s = s.replace('᾿Ο','Ὀ')
    s = s.replace('᾿Υ','᾽Υ')
    s = s.replace('᾿Ω','Ὠ')
    s = s.replace(' ·','·')
    s = s.replace('&gt;;','&gt; ;')
    s = re.sub('(δ᾿|δι᾿)',"δ’",s)
    s = re.sub('(ἀλλ᾿)',"ἀλλ’",s)
    s = re.sub('•','·',s)
    s = re.sub('δι᾿',"δι’",s)
    s = re.sub('([0-9])(al|ss|sq)','\g<1> \g<2>',s)
    s = re.sub('\s+([,;])','\g<1>',s)
    s = re.sub('–','—',s)
    s = re.sub('〉','&gt;',s)
    s = re.sub('〈','&lt;',s)
    s = re.sub('—;','— ;',s)
    s = re.sub('\);',') ;',s)
    s = re.sub('α̣','ᾳ',s)
    s = re.sub('ἢ','ἢ',s)
    s = re.sub('ῷͅ','ῷ',s)
    s = re.sub('χ;χ','χ',s)
    s = re.sub(' ʼ ',' ',s)
    s = re.sub('%2','',s)
    s = re.sub('παῤ',"παρ'",s)
    s = re.sub("᾿","'",s)
    s = re.sub('Ἔ','Ἔ',s)
    
    return(s)
    


def dowork(workn):
    englabel = ''
    grclabel = ''
    curvolume = ''
    if(workn < 66):
        editor = 'kilburn'
        curmeta = '<ti:description xml:lang="eng">Lucian, Vol. 6. Kilburn, K., editor. London: William Heinemann, Ltd.; Cambridge, MA: Harvard University Press, 1959.</ti:description>'

        archivescan = 'https://archive.org/details/luciankilb06luciuoft/page/PNUM/mode/2up'
        #https://hdl.handle.net/2027/osu.32435024202095?urlappend=%3Bseq=50%3Bownerid=13510798903769797-56
        hathiscan = 'https://hdl.handle.net/2027/osu.32435024202103?urlappend=%3Bseq=PNUM'
        hathioffset = 16
    else:
        editor = 'macleod'
        curmeta = '<ti:description xml:lang="eng">Lucian, Vol. 7. Macleod, M. D., editor. London: William Heinemann, Ltd.; Cambridge, MA: Harvard University Press, 1961.</ti:description>'
        hathiscan = 'https://hdl.handle.net/2027/osu.32435024202095?urlappend=%3Bseq=PNUM'
        archivescan = 'https://archive.org/details/lucianmac07luciuoft/page/PNUM/mode/2up'
        hathioffset = 16
    curwork = 'tlg0062.tlg0' + str(workn)
    
    curengmeta = re.sub(' editor\.',' translator.',curmeta)
    
    fname = curwork + '.' + editor + '_eng.xml'
    f = open(fname)
    for l in f:
        l = re.sub('\s+$','',l)
        m = re.search('<biblScope unit="volume">([0-9]+)',l)
        if(m):
            curvolume = m[1]
        m = re.search('<title>([^<]+)',l)
        if(m and not englabel):
            englabel ='<ti:label xml:lang="eng">' +  m[1]  + '</ti:label>'
    
        m = re.search('<title xml:lang="grc">([^<]+)',l)
        if(m and not englabel):
            grclabel = '<ti:label xml:lang="grc">' + m[1] + '</ti:label>'
        m = re.search('<pb n="(v\.[0-9]\.p\.|[0-9]+\.|)([0-9]+)"',l)
        if(m):
            curpage = m[2]
            if(not curwork in firstpages):
                firstpages[curwork] = int(curpage)
                #print('firstpage',curwork,firstpages[curwork])
    
    f.close()
    
    sourcefname = '/Users/gcrane/github/canonical-greekLit/data/tlg0062/tlg0' + str(workn) + '/__cts__.xml'
    sourcef = open(sourcefname)
    outfname = '/Users/gcrane/github/first1kwork/data/tlg0062/tlg0' + str(workn) + '/__cts__.xml'
    outf = open(outfname,'w')
    
    editionurn = 'urn:cts:greekLit:'+curwork+'.perseus-grc3'
    edition = '<ti:edition urn="'+editionurn+'" workUrn="urn:cts:greekLit:'+curwork+'" xml:lang="grc">'       
    translation = '<ti:translation urn="urn:cts:greekLit:'+curwork+'.perseus-eng3" workUrn="urn:cts:greekLit:'+curwork+'" xml:lang="eng">'       

    for l in sourcef:
        l = re.sub('\s+$','',l)
        

        if(re.search('</ti:work>',l)):
            break
        print(l,file=outf)

    print('\n' + edition,file=outf)
    print(grclabel,file=outf)
    print(curmeta,file=outf)
    print('</ti:edition>\n',file=outf)

    print('\n'+translation,file=outf)
    print(englabel,file=outf)
    print(curengmeta,file=outf)
    print('</ti:translation>\n',file=outf)
    print('</ti:work>',file=outf)
    
    sourcef.close()
    outf.close()

    
    ## copy the Greek text
    outfname = '/Users/gcrane/github/first1kwork/data/tlg0062/tlg0' + str(workn) + '/'+curwork+'.perseus-grc3.xml'
    fname = curwork + '.' + editor + '_grc.xml'

    f = open(fname)
    outf = open(outfname,'w')
    curbook = ''
    cursection = ''
    #print('edit',editionurn)
    for l in f:
        l = re.sub('\s+$','',l)
        l = regacc(l)
        
        if(re.search('<ref target="https://(ryanfb|archive|hdl.handle)',l)):
            curpage = firstpages[curwork]
            archlink = re.sub('PNUM',str(curpage),archivescan)
            print('<ref target="'+archlink+'">Internet Archive</ref>',file=outf)
            print('<ref target="'+archlink+'">Internet Archive</ref>')
            hathipage = str(curpage + hathioffset)
            htlink = re.sub('PNUM',hathipage,hathiscan)
            print('<ref target="'+htlink+'">Hathi Trust</ref>',file=outf)
            continue
        
        m = re.search('<div[^>]+subtype="book" n="([^"]+)"',l)
        if(m):
            curbook = ':' + m[1]
            l = re.sub('(<div[^>]+subtype="book")\s+','\g<1> xml:base="'+editionurn+'" ',l)
        
        m = re.search('<div[^>]+subtype="section" n="([^"]+)"',l)
        if(m):
            l = re.sub('(<div[^>]+subtype="section")\s+','\g<1> xml:base="'+editionurn+curbook+'" ',l)
        
        l = re.sub('<editor>Kenneth Kilburn[^<]+','<editor>Kenneth Kilburn',l)
        if(re.search('(edition" n=")[^"]+',l)):
            l = re.sub('(edition" n=")[^"]+','\g<1>'+editionurn,l)
            #print(l)
        l = re.sub('<note','<note xml:lang="mul"',l)
        if(curvolume):
            l = re.sub('<pb n="([0-9]+)"','<pb n="'+curvolume+'.\g<1>"',l)
        l = re.sub('<pb n="([0-9]+)\.([0-9]+)"','<pb n="v.\g<1>.p.\g<2>"',l)
        print(l,file=outf)
    
    ## now copy the English translation

    outfname = '/Users/gcrane/github/first1kwork/data/tlg0062/tlg0' + str(workn) + '/'+curwork+'.perseus-eng3.xml'
    f = open(re.sub('_grc','_eng',fname))
    outf = open(outfname,'w')
    engurn = re.sub('grc','eng',editionurn)
    curbook = ''
    cursect = ''
    for l in f:
        
        l = re.sub('\s+$','',l)
        
        
        if(re.search('<ref target="https://(ryanfb|archive|hdl.handle)',l)):
            curpage = firstpages[curwork]
            archlink = re.sub('PNUM',str(curpage),archivescan)
            print('<ref target="'+archlink+'">Internet Archive</ref>',file=outf)
            print('<ref target="'+archlink+'">Internet Archive</ref>')
            hathipage = str(curpage + hathioffset)
            htlink = re.sub('PNUM',hathipage,hathiscan)
            print('<ref target="'+htlink+'">Hathi Trust</ref>',file=outf)
            continue
        
        l = re.sub('<editor>Kenneth Kilburn[^<]+','<editor role="translator">Kenneth Kilburn',l)
        l = re.sub('<editor>Matthew Donald Macleod','<editor role="translator">Matthew Donald Macleod',l)
        m = re.search('<div[^>]+subtype="book" n="([^"]+)"',l)
        if(m):
            curbook = ':' + m[1]
            l = re.sub('(<div[^>]+subtype="book")\s+','\g<1> xml:base="'+engurn+'" ',l)
        
        m = re.search('<div[^>]+subtype="section" n="([^"]+)"',l)
        if(m):
            l = re.sub('(<div[^>]+subtype="section")\s+','\g<1> xml:base="'+engurn+curbook+'" ',l)
        if(re.search('(edition" n=")[^"]+',l)):
            l = re.sub('(edition" n=")[^"]+','translation" n="'+engurn,l)
            l = re.sub('grc"','eng"',l)
            #print('engurn',engurn,l)
            
            
        if(curvolume):
            l = re.sub('<pb n="([0-9]+)"','<pb n="'+curvolume+'.\g<1>"',l)
        l = re.sub('<pb n="([0-9]+)\.([0-9]+)"','<pb n="v.\g<1>.p.\g<2>"',l)
        
        l = re.sub('<note','<note xml:lang="eng"',l)
    
        print(l,file=outf)
    f.close()
    outf.close()
   


    return
    
    print('</ti:work>',file=outf)
    
    sourcef.close()
    outf.close()
    

for i in range(53,70):
    #dowork(i,'grc')
    dowork(i)
    
```
