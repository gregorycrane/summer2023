```python
import re

curdiv = 0
f = open('emilysmith.xml')
for l in f:
    if(re.search('<div ',l)):
        curdiv = curdiv + 1
    if(re.search('</div>',l)):
        curdiv = curdiv - 1
    if(re.search('tlg0062',l)):
        print(curdiv,l)
    if(re.search('"section"',l)):
        print(curdiv,l)

f.close()
```
