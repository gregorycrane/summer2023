```python
import re

basefile = 'lucian-dindorf'
finput = basefile + '.txt'
f = open(finput)

foutput = basefile + '-out.txt'
goutput = basefile + '-grc1.txt'
eoutput = basefile + '-eng1.txt'
outf = open(foutput,'w')
outfg = open(goutput,'w')
outfl = open(eoutput,'w')

def islatin(word):
    if(re.search('[a-zA-Z]',word)):
        return(1)
    return(0)

def isgreek(word):
    for foo in word:
        if(foo in chartable):
            #print("greek",foo,word)
            return(1,foo)
    return(0,'')
bigsents = {}


def addbigsent(s):
    global bigsents
    
    curlen = len(s.split())
    if (not curlen in bigsents):
        bigsents[curlen] = []
    bigsents[curlen].append(s)

curpage = ''
badlines = {}
totbad = 0
curscan = '0'

for l in f:
    prevword = ''
    l = re.sub('\s+$','',l)
    l = re.sub('^[\|¦\s]+','',l)
    l = re.sub('ſ','f',l)
    l = re.sub('æ','ae',l)
    l = re.sub('o([^a-zA-Z]*[α-ω])','ο\g<1>',l)
    if(re.search('^#g(title|work)',l)):
        print('\n'+l,file=outfg)
        continue
    if(re.search('^#l(title|work)',l)):
        print('\n'+l,file=outfl)
        continue
    m = re.search('## p. ([0-9]+)\s+\(#([0-9]+)',l)
    if(m):
        curpage = m[1]
        curscan = m[2]
        print('\n\n<pb n="'+curpage+'"/>',file=outf)
        print('<ref n="scan:'+m[2]+'"/>',file=outf)
        print('\n\n<pb n="'+curpage+'"/>',file=outfg)
        print('<ref n="scan:'+m[2]+'"/>',file=outfg)
        print('\n\n<pb n="'+curpage+'"/>',file=outfl)
        print('<ref n="scan:'+m[2]+'"/>',file=outfl)
        continue
    if(int(curscan)>836):
        continue
    subs1 = '^([^a-z]+[α-ωΑ-ΩΣἡἢὁςὦήἃᾧ ᾖὉ][^ a-zA-Z]*[«\.\|¦\[\s\]]+)(\(*[a-zA-Z][^α-ω|]+)$'
    junks = '\s*[\|\}]\s*$'
    m = re.search('^([^a-z]+[α-ωΣἡἢὁςὦή\—\-][^ ]*)[\|\s]+([0-9]+\.\s+[A-Za-z][^α-ω]+)$',l)
    if(m):
    #if(re.search('^([^a-z]+[α-ωΣἡἢὁςὦή\—\-][^ ]*)[\|\s]+([0-9]+\.\s+[A-Za-z][^α-ω]+)$',l)):
        #print('ref',curpage,l)
        print(re.sub(junks,'',m[1]),file=outfg)
        print(re.sub(junks,'',m[2]),file=outfl)
        #l = re.sub('^([^a-z]+[α-ωΣἡἢὁςὦή\—\-][^ ]*)[\|\s]+([0-9]+\.\s+[A-Za-z][^α-ω]+)$','#GRC2:\g<1>\n#LAT:\g<2>',l)
        if(curpage in badlines):
            badlines[curpage] = badlines[curpage] + 1
            totbad = totbad + 1
        else:
            badlines[curpage] = 1
            totbad = totbad + 1
        print(l,file=outf)
        continue

    elif(re.search(subs1,l)):
        m = re.search(subs1,l)
        #print('l1',l,file=outf)
        l = re.sub(subs1,'#GRC:\g<1>\n#LAT:\g<2>',l)
        totbad = totbad + 1
        if(curpage in badlines):
            badlines[curpage] = badlines[curpage] + 1
        else:
            badlines[curpage] = 1
        print(l,file=outf)
        print(re.sub(junks,'',m[1]),file=outfg)
        print(re.sub(junks,'',m[2]),file=outfl)
        continue
        #print(l)
    if(not re.search('#GRC',l) and len(l.split())>15):
        print('big',curpage,l)
        addbigsent(l)
    if(re.search('[a-z][a-z][a-z]',l)):
        print('#LAT3:'+l,file=outf)
        print(re.sub(junks,'',l),file=outfl)
    elif(re.search('[α-ωἡἢὁςὦήἃᾧᾖὉ]',l)):
        print('#GRC3:'+l,file=outf)
        print(re.sub(junks,'',l),file=outfg)
    else:
        print(l,file=outf)
    
    continue
f.close()
outf.close()
outfg.close()
outfl.close()

sofar = 0

print(totbad)
i = 0
for foo in sorted(badlines,key=badlines.get,reverse=True):
    i = i + 1
    sofar = badlines[foo] + sofar
    print(i,sofar,foo,badlines[foo])
```
