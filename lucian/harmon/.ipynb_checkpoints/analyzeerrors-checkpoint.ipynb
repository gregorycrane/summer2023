```python
import re
from greek_accentuation.characters import *
from greek_accentuation.syllabify import *
from greek_normalisation.utils import *


errors = {}

errortypes = ['nodot','deelis','accent','misc']

totwords = 0
totchars = 0
totcorrs = 0

for foo in errortypes:
    errors[foo] = 0
def proctrans(wknum):
    global totcorrs
    global totwords
    global totchars
    workwords = 0
    workcorrs = 0
    if(wknum < 10):
        wknums = '0' + str(i)
    else:
        wknums = str(i)
    fname = 'gk/corr-tlg0062.tlg0' + wknums  + '.perseus-grc2.xml'
    f = open(fname)
    intext = 0
    for l in f:
        res = re.findall('(.{0,20})<choice><sic>([^<]+)</sic>\s*<corr>([^<]+)</corr></choice>(.{0,20})',l)
        if(re.search('<text',l)):
            intext = 1
        l = re.sub('<note[^>]+>.+</note>',' ',l)
        if(not intext):
            continue
        workcorrs = len(res) + workcorrs
        for foo in res:
            prevt,sic, corr,postt = foo
            errclass = 'misc'
            if(sic == re.sub('·','',corr)):
                errclass = 'nodot'
                errors['nodot'] = errors['nodot'] + 1
            elif( strip_accents(sic) == strip_accents(corr)):
                errors['accent'] = errors['accent'] + 1
                errclass = 'accent'

            elif(sic == 'δʼ' and corr == 'δὲ'):
                errclass = 'deelis'
            errors[errclass] = errors[errclass] + 1
                
            print(errclass,prevt,sic,corr,postt,sep="\t")
        workl = re.sub('<[^>]+>',' ',l)
        totchars = totchars + len(workl)
        workwords = workwords + len(workl.split())
        
    totwords = totwords + workwords
    totcorrs = totcorrs + workcorrs
    print(workwords,workcorrs,totwords,totcorrs,fname,(totwords-errors['misc'])/totwords)
    
    
    f.close()
    
for i in range(1,53):
    proctrans(i)

for foo in errors:
    print(foo,errors[foo])
    
print('char acc',totchars,totcorrs,(totchars-totcorrs)/totchars)
```
