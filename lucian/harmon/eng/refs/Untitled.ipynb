```python
import re

def additem(curitem,curdict):
    if(curitem in curdict):
        curdict[curitem] = curdict[curitem] + 1
    else:
        curdict[curitem] = 1

def dofile(fnum):
    refs = {}
    notes = {}
    notetexts = {}
    fstr = str(fnum)
    
    fname = 'lucian-loeb0'+fstr+'.1st1k-eng1.xml'
    outfname = '../lucian-loeb0'+fstr+'.1st1k-eng1.xml'
    
    f = open(fname)
    for l in f:
        l = re.sub('\s+$','',l)
        res = re.findall('<ref n="([^"]+)"',l)
        for foo in res:
            additem(foo,refs)
            
        m = re.search('<note n="([^"]+)">(.+)</note>',l)
        if(m):
            additem(m[1],notes)
            if(notes[m[1]]>1):
                print('already used!',m[1])
            notetexts[m[1]] = m[2]

        
        
    
    f.close()
    for foo in refs:
        if(not foo in notes):
            print(fnum,'note missing',foo)
    for foo in notes:
        if(not foo in refs):
            print(fnum,'ref missing',foo)
    f = open(fname)
    outf = open(outfname,'w')
    for l in f:
        l = re.sub('\s+$','',l)
        while(re.search('<ref n="([^"]+)"/>',l)):
            m = re.search('<ref n="([^"]+)"/>',l)
            curcit = m[1]
            if(not curcit in notes):
                print('no notes!',curcit)
            else:
                l = re.sub('<ref n="([^"]+)"/>','<note n="\g<1>">'+notetexts[curcit]+'</note>',l,1)
        if(re.search('^<note ',l)):
            continue
            
        l = re.sub('<note n="([0-9]+)\.([0-9]+)"','<note n="v.'+str(fnum)+'.p.\g<1>.n.\g<2>"',l)
        l = re.sub('<pb n="([0-9]+)"','<pb n="v.'+str(fnum)+'.p.\g<1>"',l)
        print(l,file=outf)
    
    outf.close()
    f.close()
    
for i in range(2,6):
    dofile(i)
```
