```python
import re

def dofile(workn):
    if(workn<10):
        works = 'tlg0062.tlg00' + str(workn)
    else:
        works = 'tlg0062.tlg0' + str(workn)
        
    fname = 'corr-' + works + '.perseus-grc2.xml'
    
    outfname = '/Users/gcrane/github/first1kwork/data/' + re.sub('\.(.+)','/\g<1>/',works) + works + '.perseus-grc2.xml'
    outf = open(outfname,'w')
    
    print(outfname)
    f = open(fname)
    for l in f:
        l = re.sub('\s+$','',l)
        l = re.sub('</*choice>','',l)
        
        l = re.sub('<sic>[^<]+</sic>','',l)
        
        
        
        l = re.sub('<editor>A\. M\. Harmon','<editor>Austin Morris Harmon, 1878-1950',l)
        l = re.sub('<editor>Austin Morris Harmon, 1878-1950','<editor>Austin Morris Harmon',l)
        if(re.search('<funder n="org:Google">',l)):
            print('<funder n="org:HarvardCL">Harvard Library Arcadia Fund</funder>',file=outf)
        if(re.search('</revisionDesc>',l)):
            print('   <change who="Gregory Crane" when="2023-07-21">(1) extracted notes from the 1st1K version of these files and added them here; (2) used access to a second version of the Greek edition to compare the two and correct errors here.</change>'
,file=outf)
        m = re.search('(<note[^>]*>)',l)
        if(m):
            if(not re.search('xml:lang',m[1])):
                
                l = re.sub('<note','<note xml:lang="mul"',l)
        print(l,file=outf)
        if(re.search('<name>Lisa Cerrato</name>',l)):
            print('<name>Gregory Crane</name>',file=outf)
        
            
    
    f.close()
    outf.close()
    
for i in range(1,53):
    dofile(i)
```
