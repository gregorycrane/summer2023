```python
import re

hitdict = {}
workdict = {}


def additem(curitem,curdict):
    if(curitem in curdict):
        curdict[curitem] = curdict[curitem] + 1
    else:
        curdict[curitem] = 1
def getcits(fname):
    
    f = open(fname)
    curhits = 0
    for l in f:
        hits = re.findall('([A-Z][a-z \.,]+)\s+<title[^>]*>([^<]+)',l)

        for hit in hits:

            if(not hit[1][0].isupper()):
                #print('skipping',hit)
                continue
            curhits = curhits + 1
            additem(hit,hitdict)
            additem(hit[1],workdict)
            #print(curhits,hit,type(hit),hit[1])
    f.close()


getcits('lucian_1888.xml')

print(len(workdict))
entries = 0
for foo in sorted(workdict,key=workdict.get,reverse=True):
    entries = entries + workdict[foo]
    print(entries,foo,workdict[foo])
```
