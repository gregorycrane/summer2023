```python
import re

worklist = {}

worklist['(Achill\. Tat\.)'] = 'tlg2133.tlg001'
worklist['(Aelian\. V[ar]*\. Hist.|V[ar]*\. Hist\.)'] = 'tlg0545.tlg001'
worklist['(Aeschin\.\s+de\s+fals[\.a]\s+leg[at]*\.)'] = 'tlg0026.tlg002'
worklist['(Aeschin\.\s+[Ii]n Ctesiph\.|[Ii]n Ctesiph\.)'] = 'tlg0026.tlg002'
worklist['(Aeschin\.\s+([Ii]n|c\.) Timarch\.|([Ii]n|c\.) Timarch\.)'] = 'tlg0026.tlg001'


worklist['(Alciphr. [Ee]pist\.|Alciphr\.)'] = 'tlg0640.tlg001'
worklist['(Anacreont\.)'] = 'tlg4150.tlg001'

worklist['(Appul\. (Metam|Met)\.)'] = 'phi1212.phi002'

worklist['(Anacreon\\b)'] = 'tlg0237'


worklist['(Aristides Or\.)'] = 'tlg0284.tlg001'
worklist['(Anthol\. (Pal|Palat)\.)'] = 'tlg7000.tlg001'

worklist['(Andoc. (de myst\.|de caede Herod\.)|(de myst\.|de caede Herod\.))'] = 'tlg0028.tlg005'
worklist['(Antiph. de caede Herod\.|de caede Herod\.)'] = 'tlg0028.tlg005'


worklist['(Arist[oph]*\.\s+Acharn\.|Acharn\.)'] = 'tlg0019.tlg001'
worklist['(Arist[oph]*\.\s+Av\.|Av\.)'] = 'tlg0019.tlg006'
worklist['(Arist[oph]*\.\s+Eccl\.|Eccl\.)'] = 'tlg0019.tlg010'
worklist['(Arist[oph]*\.\s+(Equ\.|Equit\.)|(Equ\.|Equit\.))'] = 'tlg0019.tlg002'

worklist['(Arist[oph]*\.\s+Lysistr\.|Lysistr\.)'] = 'tlg0019.tlg007'
worklist['(Arist[oph]*\.\s+Nub\.|Nub\.)'] = 'tlg0019.tlg003'
worklist['(Arist[oph]*\.\s+Ran\.|Ran\.)'] = 'tlg0019.tlg009'
worklist['(Arist[oph]*. Pac\.|Pac\.)'] = 'tlg0019.tlg005'
worklist['(Arist[oph]*. Plut\.)'] = 'tlg0019.tlg011'
worklist['(Arist[oph]*\.\s+Thesm[oph]*\.|Thesm[oph]*\.)'] = 'tlg0019.tlg008'
worklist['(Arist[oph]*\.\s+Vesp\.|Vesp\.)'] = 'tlg0019.tlg004'



worklist['(Arist[ot]*\.\s+de republ\.)'] = 'tlg0086.tlg003'
worklist['(Arist[ot]*\.\s+(Eth|Ethic)\.\s+(Nic|Nicom)\.)'] = 'tlg0086.tlg010'
worklist['(Arist[ot]*\.\s+(Eth|Ethic)\.\s+(Eud|Eudem)\.|(Eth|Ethic)\.\s+(Eud|Eudem)\.)'] = 'tlg0086.tlg009'
worklist['(Arist[ot]*\.\s+Probl\.)'] = 'tlg0086.tlg036'

worklist['(Arist[ot]*\.\s+Poet\.)'] = 'tlg0086.tlg034'
worklist['(Arist[otle\.s\']+\s+(Politics|Polit\.))'] = 'tlg0086.tlg035'
worklist['(Arrian. Dissert. Epictet\.)'] = 'tlg0557.tlg001'



worklist['(Artemid. Onirocr\.)'] = 'tlg0553.tlg001'


worklist['(Athen\.|Ath\.)'] = 'tlg0008.tlg001'
worklist['(Athenaeus)'] = 'tlg0008'

worklist['(Basil. Hexaëm. homil)'] = 'tlg2040.tlg001'


worklist['(Callim. Hymn. in Jov\.)'] = 'tlg0533.tlg002:1'


worklist['(Charit\.)'] = 'tlg0554.tlg001'


worklist['(Cic. ad fam\.)'] = 'phi0474.phi056'
worklist['(Cic. de amicit\.)'] = 'phi0474.phi051'
worklist['(Cic. Brut\.)'] = 'phi0474.phi039'
worklist['(Cic. [Dd]ivin\.)'] = 'phi0474.phi053'
worklist['(Cic. in Pison\.)'] = 'phi0474.phi027'
worklist['(Cic. (Offic|Off)\.)'] = 'phi0474.phi055'
worklist['(Cic. (de orat\.)|(de orat\.))'] = 'phi0474.phi037'
worklist['(Cic. (Tuscul|Tusc)\.)'] = 'phi0474.phi049'
worklist['(Cic. (Verr|Verrin)\.)'] = 'phi0474.phi005'



worklist['(Corp\.\s+Inscr\.)'] = 'EPIG'

worklist['(Demosth\.\s+(adv\.|in|c\.)\s+Aristocr\.)'] = 'tlg0014.tlg023'
worklist['(Demosth\.\s+adv\.\s+Boet.\s+<foreign xml:lang="grc">ὀνόμ</foreign>\.|(in|adv\.|c\.) Boeot. de nomine)'] = 'tlg0014.tlg039'
worklist['(Demosth\.\s+adv\.\s+Nicostr\.)'] = 'tlg0014.tlg053'
worklist['(Demosth\.\s+adv\.\s+Phaenipp\.)'] = 'tlg0014.tlg042'
worklist['(Demosth\.\s+de\s+fals[\.a]\s+leg[at]*\.)'] = 'tlg0014.tlg019'
worklist['(Demosth\.\s+de\s+cor\.|de\s+cor\.)'] = 'tlg0014.tlg018'
worklist['(Demosth\.\s+de\s+(dot\.|dote)|de\s+(dot\.|dote))'] = 'tlg0014.tlg040'
worklist['(Demosth\.\s+(adv\.|in|c\.)\s+(Conon\.|Cononem)|(adv\.|in|c\.)\s+(Conon\.|Cononem))'] = 'tlg0014.tlg054'
worklist['(Demosth\.\s+(adv\.|in|c\.)\s+(Lacr\.|Lacritum)|(adv\.|in|c\.)\s+(Lacr\.|Lacritum))'] = 'tlg0014.tlg035'
worklist['(Demosth\.\s+(adv\.|in|c\.)\s+(Pantaen\.|Pantaenetum)|(adv\.|in|c\.)\s+(Pantaen\.|Pantaenetum))'] = 'tlg0014.tlg037'

worklist['(Demosth\.\s+Epitaph\.)'] = 'tlg0014.tlg060'
worklist['(Demosth\.\s+(c\.|in|adv\.) Eubulid\.|(adv\.|[Ii]n) Eubulid\.|c\. Eubulid\.)'] = 'tlg0014.tlg057'
worklist['(Demosth\.\s+(c\.|in|adv\.) (Mid\.|Midiam)|[Ii]n (Mid\.|Midiam)|c\. (Mid\.|Midiam))'] = 'tlg0014.tlg021'
worklist['(Demosth\.\s+in Neaer\.|[Ii]n Neaer\.)'] = 'tlg0014.tlg059'
worklist['(Demosth\.\s+de [Pp]ace\\b)'] = 'tlg0014.tlg005'
worklist['(Demosth\.\s+in Aristocr\.|[Ii]n Aristocr\.)'] = 'tlg0014.tlg023'
worklist['(Demosth\.\s+in Timoth\.|[Ii]n Timoth\.)'] = 'tlg0014.tlg049'
worklist['(Demosth\.\s+in Spud\.|[Ii]n Spud\.)'] = 'tlg0014.tlg041'
worklist['(Demosth\.\s+(in|c\.) Zenoth\.|(in|c\.) Zenoth\.)'] = 'tlg0014.tlg032'
worklist['(Demosth\.\s+Phil. 1\\b)'] = 'tlg0014.tlg004'

worklist['(Dio\s+Chrysost\.\s+[oO]r\.|Chrysost. Or\.|Dio\s+Chrysost\.)'] = 'tlg0612.tlg001'

worklist['(Dio Chrysostom)'] = 'tlg0612.tlg001'



worklist['((Diodor|Diod)\.\s+Sic\.|Diodor\.|Diod\.)'] = 'tlg0060.tlg001'
worklist['((Diog|Diogen)\.\s+Laërt\.)'] = 'tlg0004.tlg001'

worklist['(Epictet. Diss\.)'] = 'tlg0557.tlg001'

worklist['(Etymol. Magn\.)'] = 'tlg4099.tlg001'

worklist['(Euripides)\\b'] = 'tlg0006'

 
worklist['((Eur|Eurip)\. (Medea|Med\.))'] = 'tlg0006.tlg003'
worklist['((Eur|Eurip)\. (Bacchae|Bacch\.))'] = 'tlg0006.tlg017'


worklist['(Eustath. ad Iliad\.)'] = 'tlg4083.tlg001'
worklist['(Eustath. ad Odyss\.)'] = 'tlg4083.tlg002'

worklist['(Gell\. Noct\. [Aa]tt\.)'] = 'phi0070.phi001'
worklist['(Harpocration)'] = 'tlg1389.tlg001'

worklist['((Heliod|Heliodor)\. Aethiop\.)'] = 'tlg0658.tlg001'
 

worklist['(Herod\.|Herodot\.)'] = 'tlg0016.tlg001'

worklist['(Heracl. Pon\.|Heracl. Pol\.)'] = 'tlg0283.tlg001'

worklist['(Hesych\.)'] = 'tlg2274.tlg001'

worklist['(Horat. Epod\.)'] = 'phi0893.phi003'
worklist['(Horat. Epist\.)'] = 'phi0893.phi005'
worklist['(Horat. [Ss]at\.)'] = 'phi0893.phi004'

worklist['(Il\.|Iliad\.)'] = 'tlg0012.tlg001'
worklist['(Odyss\.|Od\.)'] = 'tlg0012.tlg002'

worklist['(Isaeus de Dicaeogr. hered\.)'] = 'tlg0017.tlg005'

worklist['((Isocrates|Isocr.)\s+(<foreign xml:lang="grc">π. ἀντιδ.</foreign>|Antid\.))'] = 'tlg0010.tlg018'
worklist['((Isocrates|Isocr.) Aegin\.|Aegin\.)'] = 'tlg0010.tlg018'
worklist['((Isocrates|Isocr.) (Areopag|Areop)\.|(Areopag|Areop))'] = 'tlg0010.tlg006'
worklist['((Isocrates|Isocr.) de pace)'] = 'tlg0010.tlg017'
worklist['((Isocrates|Isocr.) Panath[en]*\.|Panathen\.)'] = 'tlg0010.tlg021'
worklist['((Isocrates|Isocr.) Paneg\.|Paneg\.|Panegyr\.)'] = 'tlg0010.tlg011'
worklist['((Isocrates|Isocr.) ad Phil\.)'] = 'tlg0010.tlg020'
worklist['((Isocrates|Isocr.) c\.\s+[Ss]oph[ist]*\.|c\.\s+[Ss]oph[ist]*\.)'] = 'tlg0010.tlg008'

worklist['(Lesbonax Protrept\.)'] = 'tlg0649.tlg002'

worklist['(Liban\. Or\.)'] = 'tlg2200'


worklist['((Luc|Lucian)\. Asin\.)'] = 'tlg0061.tlg001'
worklist['((Luc|Lucian)\.\s+Amor\.)'] = 'tlg0061.tlg002'
worklist['((Luc|Lucian)\.\s+Dial. (meretr|mer)\.|Dial. (meretr|mer)\.)'] = 'tlg0062.tlg069'
worklist['((Luc|Lucian)\.\s+Deor. Dial\.|Deor. Dial.|(Luc|Lucian)\.\s+Dial. Deor\.)'] = 'tlg0062.tlg068'

 

worklist['((Luc|Lucian)\.*\s+(Alex|Alexand. s. Pseudom)\.)'] = 'tlg0061.tlg006'
worklist['((Luc|Lucian)\.*\s+[Dd]e calumn\.)'] = 'tlg0062.tlg013'
worklist['((Luc|Lucian)\.*\s+Catapl\.)'] = 'tlg0062.tlg016'
worklist['((Luc|Lucian)\.*\s+Cyn\.)'] = 'tlg0062.tlg016'
worklist['((Luc|Lucian)\.*\s+(Demon\.|Dem. enc\.))'] = 'tlg0062.tlg008'
worklist['((Luc|Lucian)\.*\s+de [Gg]ymn\.)'] = 'tlg0062.tlg034'
worklist['((Luc|Lucian)\.*\s+Hermot[im]*\.)'] = 'tlg0062.tlg063'
worklist['((Luc|Lucian)\.*\s+Hist. conscr\.)'] = 'tlg0062.tlg053'
worklist['((Luc|Lucian)\.*\s+Icaromen\.)'] = 'tlg0062.tlg021'
worklist['((Luc|Lucian)\.*\s+Imag\.)'] = 'tlg0062.tlg039'
worklist['((Luc|Lucian)\.*\s+adv. indoct\.|adv. indoct\.)'] = 'tlg0062.tlg028'
worklist['((Luc|Lucian)\.*\s+Jup\.\s+Trag\.)'] = 'tlg0062.tlg018'
worklist['((Luc|Lucian)\.*\s+(de luct\.|de luctu)|(de luct\.|de luctu))'] = 'tlg0062.tlg058'
worklist['((Luc|Lucian)\.*\s+(de lapsu in (salut|sal))\.|(de lapsu in (sal|salut))\.)'] = 'tlg0062.tlg058'
worklist['((Luc|Lucian)\.*\s+(Lexiph|Lexiphan)\.|(Lexiph|Lexiphan)\.)'] = 'tlg0062.tlg046'
worklist['((Luc|Lucian)\.*\s+de merc. cond\.)'] = 'tlg0062.tlg033'
worklist['((Luc|Lucian)\.*\s+(Navig|Nav)\.)'] = 'tlg0062.tlg065'
worklist['((Luc|Lucian)\.*\s+Necyom\.)'] = 'tlg0062.tlg035'
worklist['((Luc|Lucian)\.*\s+Nigrin\.)'] = 'tlg0062.tlg007'
worklist['((Luc|Lucian)\.*\s+([Pp]aras\.|de (parasito|paras\.)))'] = 'tlg0062.tlg030'
worklist['((Luc|Lucian)\.*\s+Philops\.)'] = 'tlg0062.tlg031'
worklist['((Luc|Lucian)\.*\s+(Pisc|Piscat)\.)'] = 'tlg0062.tlg025'
worklist['((Luc|Lucian)\.*\s+Prom. in verb\.)'] = 'tlg0062.tlg064'
worklist['((Luc|Lucian)\.*\s+Quomod. hist. conscr\.|Quomod. hist. conscr\.)'] = 'tlg0062.tlg053'

worklist['((Luc|Lucian)\.*\s+[Ee]pist\.\s+Saturn\.)'] = 'tlg0062.tlg055'
worklist['((Luc|Lucian)\.*\s+Rhet. praec\.|Rhet. praec\.)'] = 'tlg0062.tlg029'
worklist['((Luc|Lucian)\.*\s+de [Ss]alt\.)'] = 'tlg0062.tlg045'
worklist['((Luc|Lucian)\.*\s+(de salt|de saltat)\.)'] = 'tlg0062.tlg045'
worklist['((Luc|Lucian)\.*\s+Saturn\.)'] = 'tlg0062.tlg055'
worklist['((Luc|Lucian)\.*\s+Soloec\.|Soloec\.)'] = 'tlg0062.tlg070'
worklist['((Luc|Lucian)\.*\s+Somn\.|Somn\.)'] = 'tlg0062.tlg029'
worklist['((Luc|Lucian)\.*\s+Symp\.)'] = 'tlg0062.tlg015'
worklist['((Luc|Lucian)\.*\s+de Syr\.)'] = 'tlg0062.tlg041'

worklist['((Luc|Lucian)\.*\s+Tim\.)'] = 'tlg0062.tlg023'
worklist['((Luc|Lucian)\.*\s+(Toxar|Tox)\.)'] = 'tlg0062.tlg044'

worklist['(Ly[ck]urg. in Leocr\.)'] = 'tlg0034.tlg001'



worklist['(Lysias de (caede|caed\.) Erat\.)'] = 'tlg0540.tlg001'
worklist['(Lysias de inval\.)'] = 'tlg0540.tlg024'
worklist['(Lysias de bonis Aristoph\.)'] = 'tlg0540.tlg019'
worklist['(Lysias in Theomn. I\\b)'] = 'tlg0540.tlg010'



worklist['((Mart\.|Martial) Epigr\.|Mart\.)'] = 'phi1294.phi002'

worklist['(Maxim. Tyr. Diss\.)'] = 'tlg0563.tlg001'

worklist['(Ovid\. Amor\.)'] = 'phi0959.phi001'
worklist['(Ovid\. Trist\.)'] = 'phi0959.phi008'

worklist['(Palaeph. Incredib\.)'] = 'tlg1553.tlg001'
worklist['(Pausan\.)'] = 'tlg0525.tlg001'
worklist['(Paroemiogr. Gr\.)'] = 'Paroemiag_Gr'

 
worklist['(Petron. Sat\.|Petron\.)'] = 'phi0972.phi001'
worklist['(Phot. (Bibl|Lex)\.)'] = 'tlg4040.tlg001'

worklist['(Philostr. Vit. Apollon. Tyan\.)'] = 'tlg0638.tlg001'

worklist['(Pindar\.\s+Isthm\.)'] = 'tlg0033.tlg004'
worklist['(Pindar\.\s+Ol\.)'] = 'tlg0033.tlg001'
worklist['((Plato.s|Plat[o.]+)\s+(Alcib\.\s+2))'] = 'tlg0059.tlg014'
worklist['((Plato.s|Plat[o.]+)\s+(Alcib\.))'] = 'tlg0059.tlg013'
worklist['((Plato.s|Plat[o.]+)\s+(Leges|Legg\.|Leg\.))'] = 'tlg0059.tlg034'
worklist['((Plato.s|Plat[o.]+)\s+Crit\.|Crit\.)'] = 'tlg0059.tlg003'
worklist['((Plato.s|Plat[o.]+)\s+(Gorg\.|Gorgias))'] = 'tlg0059.tlg023'
worklist['((Plato.s|Plat[o.]+)\s+(Lys\.|Lysis))'] = 'tlg0059.tlg020'
worklist['((Plato.s|Plat[o.]+)\s+Menex\.|Menex\.)'] = 'tlg0059.tlg028'
worklist['((Plato.s|Plat[o.]+)\s+Meno|Meno)'] = 'tlg0059.tlg024'
worklist['((Plato.s|Plat[o.]+)\s+Phaed\.|Phaed\.)'] = 'tlg0059.tlg004'
worklist['((Plato.s|Plat[o.]+)\s+Phaedr\.|Phaedr\.)'] = 'tlg0059.tlg012'
worklist['((Plato.s|Plat[o.]+)\s+Phileb\.|Phileb\.)'] = 'tlg0059.tlg010'
worklist['((Plato.s|Plat[o.]+)\s+Protag\.|Protag\.)'] = 'tlg0059.tlg022'
worklist['((Plato.s|Plat[o.]+)\s+(de republ\.|de republ\.|Republ\.))'] = 'tlg0059.tlg030'
worklist['((Plato.s|Plat[o.]+)\s+Theaet[et]*\.|Theaet[et]*\.)'] = 'tlg0059.tlg006'
worklist['((Plato.s|Plat[o.]+)\s+Tim\.|Timaeus)'] = 'tlg0059.tlg006'
worklist['(Plato\s+Republ\.)'] = 'tlg0059.tlg030'

worklist['(Pseudo-Plat. Axioch\.)'] = 'tlg0059.Axiochus'
 
    
worklist['(Plaut\. Aulul\.)'] = 'phi0119.phi003'
worklist['(Plaut\. Asin\.)'] = 'phi0119.phi002'

worklist['(Plaut\. Bacch[id]*\.)'] = 'phi0119.phi004'
worklist['(Plaut\. Capt\.)'] = 'phi0119.phi005'
worklist['(Plaut\. Casin\.)'] = 'phi0119.phi006'
worklist['(Plaut\. (Cistell|Cist)\.|Cistell\.|Cist\.)'] = 'phi0119.phi007'
worklist['(Plaut\. Epid\.|Epid\.)'] = 'phi0119.phi009'
worklist['(Plaut\. Merc\.)'] = 'phi0119.phi011'
worklist['(Plaut\. Mil. glor\.|Mil. glor)'] = 'phi0119.phi012'
worklist['(Plaut\. Mostell\.|Mostell)'] = 'phi0119.phi013'
worklist['(Plaut\. Pseud\.|Pseud\.)'] = 'phi0119.phi016'
worklist['(Plaut\. Trin\.|Trin\.)'] = 'phi0119.phi019'
worklist['(Plaut\. Rudens\\b)'] = 'phi0119.phi017'


worklist['(Plin\. N\.\s*H\.)'] = 'phi0978.phi001'


worklist['(Plut[arch]*. adv. Stoic\.|adv. Stoic\.)'] = 'tlg0007.tlg138'
worklist['(Plut[arch]*. Amator\.\s+narr\.|Amator\.\s+narr\.)'] = 'tlg0007.tlg114'
worklist['(Plut[arch]*. Apophthegm. Alex\.|Apophthegm. Alex\.)'] = 'tlg0007.tlg0??'
worklist['(Plut[arch]*. Apophth. Lacon\.|Apophth. Lac[on]*\.)'] = 'tlg0007.tlg082'
worklist['(Plut[arch]*. de cohib\. ira\\b|de cohib\. ira\\b)'] = 'tlg0007.tlg095'
worklist['(Plut[arch]*. adv. Colot\.|adv. Colot\.)'] = 'tlg0007.tlg140'
worklist['(Plut[arch]*. (de audit\.|de audit\.|de aud\.))'] = 'tlg0007.tlg069'
worklist['(Plut[arch]*. conv. sept\.)'] = 'tlg0007.tlg102'
worklist['(Plut[arch]*. Consol. ad Apoll\.)'] = 'tlg0007.tlg076'
worklist['(Plut[arch]*. de curios\.|de curios\.)'] = 'tlg0007.tlg102'
worklist['(Plut[arch]*. (de educ\. puer|educ|de educ)\.|de educ\.)'] = 'tlg0007.tlg067'
worklist['(Plut[arch]*. (fort|de fort)\.|de fort\.)'] = 'tlg0007.tlg074'
worklist['(Plut[arch]*. de Isid. et Osir\.|de Isid. et Osir\.)'] = 'tlg0007.tlg089'


worklist['(Plut[arch]*. Reip. gerend. praec\.|Reip. gerend. praec\.)'] = 'tlg0007.tlg118'
worklist['(Plut[arch]*. De san. praecept\.)'] = 'tlg0007.tlg077'
worklist['(Plut[arch]*. Inst\. Lac\.)'] = 'tlg0007.tlg0??'
worklist['(Plut[arch]*. Defect. [Oo]rac\.|Defect. Orac\.)'] = 'tlg0007.tlg092'
worklist['(Plut[arch]*. de rect\. rat\.|rect\. rat\.)'] = 'tlg0007.tlg069'
worklist['(Plut[arch]*. de ser. num. vind\.|de ser. num. vind\.)'] = 'tlg0007.tlg107'
worklist['(Plut[arch]*. de exilio|de exilio)'] = 'tlg0007.tlg110'
worklist['(Plut[arch]*. de Herodoti malign|de Herodoti malign)'] = 'tlg0007.tlg123'
worklist['(Plut[arch]*. de superst\.)'] = 'tlg0007.tlg080'
worklist['(Plut[arch]*. de vitios. pud\.|de vitios. pud\.)'] = 'tlg0007.tlg104'
worklist['(Plut[arch]*. Gryll\.)'] = 'Plutarch-Gryllus'
worklist['((Pseudo-Plut|Plut)[arch]*. (X [oO]rat. vit\.|Vit. X Orat\.)|X [oO]rat. vit\.|Vit. X Orat\.)'] = 'tlg0007.tlg121'
worklist['(Plut[arch]*. Quaest. [Gg]raec\.|Quaest. [Gg]raec\.)'] = 'tlg0007.tlg084b'
worklist['(Plut[arch]*. (Symp\.|Stoic. rep\.|[Dd]e Stoic. repugn\.)|[Dd]e Stoic. (repugn|rep)\.)'] = 'tlg0007.tlg136'
worklist['(Plut[arch]*. (Symp\.|Sept. sap. conv\.)|Sept. sap. conv\.)'] = 'tlg0007.tlg079'




worklist['(Plut[arch’s]*. (Ages|Agesil)\.)'] = 'tlg0007.tlg044'
worklist['(Plut[arch’s]*. (Agid\.|Agis))'] = 'tlg0007.tlg051'
worklist['(Plut[arch’s]*. (Alcib|Alc)\.)'] = 'tlg0007.tlg015'
worklist['(Plut[arch’s]*. (Aris|Aristid)\.)'] = 'tlg0007.tlg024'
worklist['(Plut[arch’s]*. Artax\.)'] = 'tlg0007.tlg064'
worklist['(Plut[arch’s]*. Apophth\.\s+reg\.)'] = 'tlg0007.tlg081'
worklist['(Plut[arch’s]*.\s+Cat\.\s+Min\.)'] = 'tlg0007.tlg025'
worklist['(Plut[arch’s]*.\s+Cim\.|Cim\.)'] = 'tlg0007.tlg035'
worklist['(Plut[arch’s]*.\s+Demetr\.)'] = 'tlg0007.tlg057'
worklist['(Plut[arch’s]*.\s+Demosth[en]*\.)'] = 'tlg0007.tlg054'
worklist['(Plut[arch’s]*.\s+Dion\.)'] = 'tlg0007.tlg060'
worklist['(Plut[arch’s]*.\s+Lyc[urg]*\.)'] = 'tlg0007.tlg004'
worklist['(Plut[arch’s]*.\s+Lys[ander]*\.)'] = 'tlg0007.tlg032'
worklist['(Plut[arch’s]*.\s+Nic\.)'] = 'tlg0007.tlg038'
worklist['(Plut[arch’s]*.\s+Pelop[id]*\.)'] = 'tlg0007.021'
worklist['(Plut[arch’s]*.\s+Pericl\.)'] = 'tlg0007.tlg012'
worklist['(Plut[arch’s]*.\s+Phoc\.)'] = 'tlg0007.tlg049'
worklist['(Plut[arch’s]*.\s+(Sol\.|Solon))'] = 'tlg0007.tlg007'
worklist['(Plut[arch’s]*.\s+Themist\.)'] = 'tlg0007.tlg010'
worklist['(Plut[arch’s]*.\s+Timol\.)'] = 'tlg0007.tlg018'



worklist['(Poll\.)'] = 'tlg0542.tlg001'

worklist['(Polyb\.|Polybius)'] = 'tlg0543.tlg001'
worklist['(Quintil\.|Quintilian)'] = 'phi1002.phi001'


worklist['(Senec. de const\.)'] = 'stoa0255.stoa009'

worklist['(Sophocl. Oed. Col\.)'] = 'tlg0011.tlg007'


worklist['(Steph\. de urbib\.)'] = 'tlg4028.tlg001'
worklist['(Stob\. Serm\.)'] = 'tlg2037.tlg001'




worklist['(Strab[o]*\.|Strabo\\b)'] = 'tlg0099.tlg001'
worklist['(Suid\.|Suidas)'] = 'tlg9010.tlg001'

worklist['(Sueton\. Vitell\.)'] = 'phi1348.abo019'

worklist['(Terent.\ Andr\.)'] = 'phi0134.phi001'
worklist['(Terent.\ Eun[uch]*\.|Eun[uch]*\.)'] = 'phi0134.phi003'
worklist['(Terent.\ Heaut\.|Heaut\.)'] = 'phi0134.phi002'
worklist['(Terent.\ Hecyr\.|Hecyr\.)'] = 'phi0134.phi005'


worklist['(Theocr\.\s+(Idyll|Id)\.|Theocr\.)'] = 'tlg1714.tlg001'

worklist['((Theophrast\.|Theophr\.)\s+(Char\.|Charact.))'] = 'tlg0093.tlg009'
worklist['((Theophrast\.|Theophr\.)\s+(Hist. plant.))'] = 'tlg0093.tlg001'
worklist['((Theophrast\.|Theophr\.)\s+(de caus. plant\.))'] = 'tlg0093.tlg002'
worklist['(Themist\. Or\.)'] = 'tlg2001'
worklist['(Thucyd\.)'] = 'tlg0003.tlg001'

worklist['(Valer. Max\.)'] = 'phi1035.phi001'
worklist['(Vitruv\.)'] = 'phi1056.phi001'



worklist['(Xenoph[ons’\.]+\s+Anab\.)'] = 'tlg0032.tlg006'
worklist['(Xenoph[ons’\.]+\s+Cyropaed\.)'] = 'tlg0032.tlg001'
worklist['(Xenoph[ons’\.]+\s+Hell\.)'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+(Oec|Oecon)\.)'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+Memor\.\s+Socr\.|Xenoph\.\s+(Memor\.|Mem\.))'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+(de republ\.|de rep. Lac\.|Reip\.))'] = 'tlg0032.tlg010'
worklist['(Xenoph[ons’\.]+\s+Symp\.)'] = 'tlg0032.tlg004'
#worklist['(Xenoph\.\s+)'] = 'tlg0032.tlg003'

def dofile(fname):
    global worklist
    sawauths = 0
    
    romarab = '([XILVC]+[,\.]\s+)([0-9\-]+,\s+[0-9\-]+|[0-9\-]+)'
    bookpage = '([XILVC]+,\s+p\.\s+[0-9\-]+)'
    pageref = '\\b(p\.\s+[0-9\-]+)'
    tmpfname = 'chartmp.xml'
    tmpf = open(tmpfname,'w')
    
    f = open(fname)
    intext = 0
    
    text = f.read()
    text = re.sub('\n+','\n',text)
    for l in text.split('\n'):
        l = re.sub('\s+$','',l)
        if(re.search('<text',l)):
            intext = 1
        if(not intext):
            print(l,file=tmpf)
            continue
            
        l = re.sub('^\s+','',l)
        l = re.sub('\s\s+',' ',l)
        l = re.sub('^\s*(<forei[^>]+>)\s*','\g<1>',l)
        l = re.sub('\s*(</foreign>)\s*$','\g<1>',l)
        l = re.sub('(<p>|<div|<p |<cit|<quote|<bibl|<note|<l)','\n\n\g<1>',l)
        l = re.sub('(<p>|<div|<p |<cit|<quote|<bibl)','\n\g<1>',l)
        l = re.sub('(</cit>|</quote>)\s*','\g<1>\n',l)
        l = re.sub('$\.\s*([0-9])','§. \g<1>',l)
        
                
            
        print(l,file=tmpf,end=' ')
    f.close()  
    tmpf.close()

    outfname = re.sub('\.xml','-out.xml',fname)
    print(outfname)
    outf = open(outfname,'w')
    f = open(tmpfname)
    for l in f:
        l = re.sub('(<quote)(>|\sxml:l)','\g<1> rend="blockquote"\g<2>',l)
        
        l = re.sub('\s+$','',l)
        l = re.sub('\\bol ','οἱ ',l)
        l = re.sub(bookpage,'<ref n="bookpage">\g<1></ref>',l)
        l = re.sub(romarab,'<ref n="romarab">\g<1>\g<2></ref>',l)
        l = re.sub(pageref,'<ref n="pageref">\g<1></ref>',l)
        l = re.sub('<foreign xml:lang="grc">π\. ἀντιδ.</foreign>','Antid.',l)
        l = re.sub('<foreign xml:lang="grc">περὶ ἀντιδ</foreign>\.','de Antid.',l)
        l = re.sub('</foreign>([^<a-zA-Z0-9]+)<foreign xml:lang="grc">','\g<1>',l)
        l = re.sub('([\s,\.;:]+[a-zA-Z][\s,\.;:a-zA-Z]+)(</foreign>)','\g<2>\g<1>',l)
        l = re.sub('(<sic>|</sic>)','',l)
        for foo in worklist:
            m = re.search(foo,l)
            if(m):
                l = re.sub(foo,'<bibl n="'+worklist[foo]+'">\g<1></bibl>',l)
                sawauths = sawauths + 1
        l = re.sub('(<bibl [^>]+>[^<]+)<bibl [^>]+>([^<]+)</bibl>','\g<1>\g<2>',l)
        l = re.sub('(<foreign xml:lang="grc">[^<A-Z:a-z]+)(\s+[:;\.\(a-z\[A-Z,0-9\s]+)','\g<1></foreign>\g<2><foreign xml:lang="grc">',l)
        
        l = re.sub('ὁ\s+([0-9]+)',' \g<1>',l)
        while(re.search('(<foreig[^>]+>[^<]*)<for[^>]+>([^<]+)</foreign>',l)):
            l = re.sub('(<foreig[^>]+>[^<]*)<for[^>]+>([^<]+)</foreign>','\g<1>\g<2>',l)
        l = re.sub('<foreign xml:lang="grc">ὁ</foreign>.\s+([0-9])','§. \g<1>',l)
        l = re.sub('<foreign xml:lang="grc">([^<]+\s+[^<]+)</foreign>','<quote xml:lang="grc">\g<1></quote>',l)
        l = re.sub('\$\.\s*([0-9]+)','§. \g<1>',l)
        print(l,file=outf)
    outf.close()
    print(sawauths)
    
#dofile('becker-text.xml')
dofile('becker-notes.xml')
#dofile('becker-appendices.xml')
```
