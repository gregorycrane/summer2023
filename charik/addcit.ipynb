```python
import re

worklist = {}

worklist['(Achill\. Tat\.)'] = 'tlg2133.tlg001'
worklist['(Aelian\. V[ar]*\. Hist.|V[ar]*\. Hist\.)'] = 'tlg0545.tlg001'
worklist['(Aeschin\.\s+de\s+fals[\.a]\s+leg[at]*\.)'] = 'tlg0026.tlg002'
worklist['(Aeschin\.\s+[Ii]n Ctesiph\.|[Ii]n Ctesiph\.)'] = 'tlg0026.tlg002'
worklist['(Aeschin\.\s+[Ii]n Timarch\.|[Ii]n Timarch\.)'] = 'tlg0026.tlg001'


worklist['(Alciphr. [Ee]pist\.|Alciphr\.)'] = 'tlg0640.tlg001'

worklist['(Appul\. (Metam|Met)\.)'] = 'phi1212.phi002'

worklist['(Anacreon\\b)'] = 'tlg0237'


worklist['(Aristides Or\.)'] = 'tlg0284.tlg001'
worklist['(Anthol\. (Pal|Palat)\.)'] = 'tlg7000.tlg001'

worklist['(Antiph. de caede Herod\.|de caede Herod\.)'] = 'tlg0028.tlg005'


worklist['(Aristoph\.\s+Acharn\.|Acharn\.)'] = 'tlg0019.tlg001'
worklist['(Aristoph\.\s+Av\.|Av\.)'] = 'tlg0019.tlg006'
worklist['(Aristoph\.\s+Eccl\.|Eccl\.)'] = 'tlg0019.tlg010'
worklist['(Aristoph\.\s+(Equ\.|Equit\.)|(Equ\.|Equit\.))'] = 'tlg0019.tlg002'

worklist['(Aristoph\.\s+Nub\.|Nub\.)'] = 'tlg0019.tlg003'
worklist['(Aristoph\.\s+Ran\.|Ran\.)'] = 'tlg0019.tlg009'
worklist['(Aristoph. Pac\.|Pac\.)'] = 'tlg0019.tlg005'
worklist['(Aristoph. Plut\.)'] = 'tlg0019.tlg011'
worklist['(Aristoph\.\s+Thesm[oph]*\.|Thesm[oph]*\.)'] = 'tlg0019.tlg008'
worklist['(Aristoph\.\s+Vesp\.|Vesp\.)'] = 'tlg0019.tlg004'



worklist['(Aristot\.\s+de republ\.)'] = 'tlg0086.tlg003'
worklist['(Aristot\.\s+(Eth|Ethic)\.\s+(Nic|Nicom)\.)'] = 'tlg0086.tlg010'
worklist['(Aristot\.\s+Probl\.)'] = 'tlg0086.tlg036'

worklist['(Aristot\.\s+Poet\.)'] = 'tlg0086.tlg034'
worklist['(Aristot[le\.s\']+\s+(Politics|Polit\.))'] = 'tlg0086.tlg035'

worklist['(Artemid. Onirocr\.)'] = 'tlg0553.tlg010'


worklist['(Athen\.|Ath\.)'] = 'tlg0008.tlg001'
worklist['(Athenaeus)'] = 'tlg0008'

worklist['(Basil. Hexaëm. homil)'] = 'tlg2040.tlg001'


worklist['(Callim. Hymn. in Jov\.)'] = 'tlg0533.tlg002:1'


worklist['(Charit\.)'] = 'tlg0554.tlg001'


worklist['(Cic. ad fam\.)'] = 'phi0474.phi056'
worklist['(Cic. Divin\.)'] = 'phi0474.phi053'
worklist['(Cic. Offic\.)'] = 'phi0474.phi055'



worklist['(Corp\.\s+Inscr\.)'] = 'EPIG'

worklist['(Demosth\.\s+adv\.\s+Boet.\s+<foreign xml:lang="grc">ὀνόμ</foreign>\.)'] = 'tlg0014.tlg039'
worklist['(Demosth\.\s+adv\.\s+Phaenipp\.)'] = 'tlg0014.tlg042'
worklist['(Demosth\.\s+de\s+fals[\.a]\s+leg[at]*\.)'] = 'tlg0014.tlg019'
worklist['(Demosth\.\s+de\s+cor\.|de\s+cor\.)'] = 'tlg0014.tlg018'

worklist['(Demosth\.\s+Epitaph\.)'] = 'tlg0014.tlg060'
worklist['(Demosth\.\s+(c\.|in|adv\.) Eubulid\.|(adv\.|[Ii]n) Eubulid\.|c\. Eubulid\.)'] = 'tlg0014.tlg057'
worklist['(Demosth\.\s+(c\.|in) Mid\.|[Ii]n Mid\.|c\. Mid\.)'] = 'tlg0014.tlg021'
worklist['(Demosth\.\s+in Neaer\.|[Ii]n Neaer\.)'] = 'tlg0014.tlg059'
worklist['(Demosth\.\s+in Timoth\.|[Ii]n Timoth\.)'] = 'tlg0014.tlg049'

worklist['(Dio\s+Chrysost\.\s+[oO]r\.|Chrysost. Or\.)'] = 'tlg0612.tlg001'

worklist['(Dio Chrysostom)'] = 'tlg0612.tlg001'



worklist['((Diodor|Diod)\.\s+Sic\.|Diodor\.|Diod\.)'] = 'tlg0060.tlg001'
worklist['(Diog\.\s+Laërt\.)'] = 'tlg0004.tlg001'

worklist['(Epictet. Diss\.)'] = 'tlg0557.tlg001'

worklist['(Etymol. Magn\.)'] = 'tlg4099.tlg001'

worklist['(Euripides)\\b'] = 'tlg0006'

 
worklist['((Eur|Eurip)\. (Medea|Med\.))'] = 'tlg0006.tlg003'
worklist['((Eur|Eurip)\. (Bacchae|Bacch\.))'] = 'tlg0006.tlg017'


worklist['(Eustath. ad Iliad\.)'] = 'tlg4083.tlg001'
worklist['(Eustath. ad Odyss\.)'] = 'tlg4083.tlg002'

worklist['(Harpocration)'] = 'tlg1389.tlg00'

worklist['((Heliod|Heliodor)\. Aethiop\.)'] = 'tlg0658.tlg001'
 

worklist['(Herod\.|Herodot\.)'] = 'tlg0016.tlg001'

worklist['(Heracl. Pon\.|Heracl. Pol\.)'] = 'tlg0283.tlg001'

worklist['(Hesych\.)'] = 'tlg2274.tlg001'

worklist['(Horat. Epod\.)'] = 'phi0893.phi003'
worklist['(Horat. Epist\.)'] = 'phi0893.phi005'
worklist['(Horat. [Ss]at\.)'] = 'phi0893.phi004'

worklist['(Il\.|Iliad\.)'] = 'tlg0012.tlg001'
worklist['(Odyss\.|Od\.)'] = 'tlg0012.tlg002'

worklist['(Isaeus de Dicaeogr. hered\.)'] = 'tlg0017.tlg005'

worklist['((Isocrates|Isocr.)\s+<foreign xml:lang="grc">π. ἀντιδ.</foreign>)'] = 'tlg0010.tlg018'
worklist['((Isocrates|Isocr.) Aegin\.|Aegin\.)'] = 'tlg0010.tlg018'
worklist['((Isocrates|Isocr.) Areopag\.)'] = 'tlg0010.tlg006'
worklist['((Isocrates|Isocr.) de pace)'] = 'tlg0010.tlg017'
worklist['((Isocrates|Isocr.) Panath[en]*\.|Panathen\.)'] = 'tlg0010.tlg021'
worklist['((Isocrates|Isocr.) Paneg\.|Paneg\.|Panegyr\.)'] = 'tlg0010.tlg011'
worklist['((Isocrates|Isocr.) ad Phil\.)'] = 'tlg0010.tlg020'
worklist['((Isocrates|Isocr.) c\.\s+[Ss]oph[ist]*\.|c\.\s+[Ss]oph[ist]*\.)'] = 'tlg0010.tlg008'

worklist['(Lesbonax Protrept\.)'] = 'tlg0649.tlg002'

worklist['(Liban\. Or\.)'] = 'tlg2200'


worklist['(Lucian. Asin\.)'] = 'tlg0061.tlg001'
worklist['(Lucian\.\s+Amor\.)'] = 'tlg0061.tlg002'
worklist['(Lucian\.\s+Dial. (meretr|mer)\.|Dial. (meretr|mer)\.)'] = 'tlg0062.tlg069'
worklist['(Lucian\.\s+Deor. Dial\.|Deor. Dial.|Lucian\.\s+Dial. Deor\.)'] = 'tlg0062.tlg068'

 

worklist['(Lucian\.*\s+Alex\.)'] = 'tlg0062.tlg038'
worklist['(Lucian\.*\s+Demon\.)'] = 'tlg0062.tlg008'
worklist['(Lucian\.*\s+Hermot[im]*\.)'] = 'tlg0062.tlg063'
worklist['(Lucian\.*\s+Icaromen\.)'] = 'tlg0062.tlg021'
worklist['(Lucian\.*\s+Jup\.\s+Trag\.)'] = 'tlg0062.tlg018'
worklist['(Lucian\.*\s+(Lexiph|Lexiphan)\.|(Lexiph|Lexiphan)\.)'] = 'tlg0062.tlg046'
worklist['(Lucian\.*\s+de merc. cond\.)'] = 'tlg0062.tlg033'
worklist['(Lucian\.*\s+Navig\.)'] = 'tlg0062.tlg065'
worklist['(Lucian\.*\s+Necyom\.)'] = 'tlg0062.tlg035'
worklist['(Lucian\.*\s+(Paras\.|de parasito))'] = 'tlg0062.tlg030'
worklist['(Lucian\.*\s+Philops\.)'] = 'tlg0062.tlg031'
worklist['(Lucian\.*\s+Pisc\.)'] = 'tlg0062.tlg025'
worklist['(Lucian\.*\s+Prom. in verb\.)'] = 'tlg0062.tlg064'
worklist['(Lucian\.*\s+Epist\.\s+Saturn\.)'] = 'tlg0062.tlg055'
worklist['(Lucian\.*\s+Somn\.)'] = 'tlg0062.tlg029'
worklist['(Lucian\.*\s+Symp\.)'] = 'tlg0062.tlg015'

worklist['(Lucian\.*\s+Tim\.)'] = 'tlg0062.tlg023'
worklist['(Lucian\.*\s+Toxar\.)'] = 'tlg0062.tlg044'

worklist['(Ly[ck]urg. in Leocr\.)'] = 'tlg0034.tlg001'



worklist['(Lysias de (caede|caed\.) Erat\.)'] = 'tlg0540.tlg001'
worklist['(Lysias de inval\.)'] = 'tlg0540.tlg024'
worklist['(Lysias de bonis Aristoph\.)'] = 'tlg0540.tlg019'
worklist['(Lysias in Theomn. I\\b)'] = 'tlg0540.tlg010'



worklist['((Mart\.|Martial) Epigr\.)'] = 'phi1294.phi002'

worklist['(Ovid\. Amor\.)'] = 'phi0959.phi001'

worklist['(Palaeph. Incredib\.)'] = 'tlg1553.tlg001'
worklist['(Pausan\.)'] = 'tlg0525.tlg001'
worklist['(Paroemiogr. Gr\.)'] = 'Paroemiag_Gr'

 
worklist['(Phot. Bibl\.)'] = 'tlg4040.tlg001'

worklist['(Philostr. Vit. Apollon. Tyan\.)'] = 'tlg0638.tlg001'

worklist['(Pindar\.\s+Isthm\.)'] = 'tlg0033.tlg004'
worklist['(Pindar\.\s+Ol\.)'] = 'tlg0033.tlg001'
worklist['((Plato.s|Plat[o.]+)\s+(Leges|Legg\.|Leg\.))'] = 'tlg0059.tlg034'
worklist['((Plato.s|Plat[o.]+)\s+Crit\.|Crit\.)'] = 'tlg0059.tlg003'
worklist['((Plato.s|Plat[o.]+)\s+Lys\.)'] = 'tlg0059.tlg020'
worklist['((Plato.s|Plat[o.]+)\s+Menex\.|Menex\.)'] = 'tlg0059.tlg028'
worklist['((Plato.s|Plat[o.]+)\s+Meno|Meno)'] = 'tlg0059.tlg024'
worklist['((Plato.s|Plat[o.]+)\s+Phaed\.|Phaed\.)'] = 'tlg0059.tlg004'
worklist['((Plato.s|Plat[o.]+)\s+Phileb\.|Phileb\.)'] = 'tlg0059.tlg010'
worklist['((Plato.s|Plat[o.]+)\s+Protag\.|Protag\.)'] = 'tlg0059.tlg022'
worklist['((Plato.s|Plat[o.]+)\s+de republ\.|de republ\.)'] = 'tlg0059.tlg030'
worklist['((Plato.s|Plat[o.]+)\s+Theaet[et]*\.|Theaet[et]*\.)'] = 'tlg0059.tlg006'
worklist['(Plato\s+Republ\.)'] = 'tlg0059.tlg030'

worklist['(Pseudo-Plat. Axioch\.)'] = 'tlg0059.Axiochus'
 
    
worklist['(Plaut\. Aulul\.)'] = 'phi0119.phi003'
worklist['(Plaut\. Asin\.)'] = 'phi0119.phi002'

worklist['(Plaut\. Bacch[id]*\.)'] = 'phi0119.phi004'
worklist['(Plaut\. Capt\.)'] = 'phi0119.phi005'
worklist['(Plaut\. Casin\.)'] = 'phi0119.phi006'
worklist['(Plaut\. Cistell\.|Cistell\.)'] = 'phi0119.phi007'
worklist['(Plaut\. Epid\.|Epid\.)'] = 'phi0119.phi009'
worklist['(Plaut\. Merc\.)'] = 'phi0119.phi011'
worklist['(Plaut\. Mil. glor\.|Mil. glor)'] = 'phi0119.phi012'
worklist['(Plaut\. Pseud\.|Pseud\.)'] = 'phi0119.phi016'
worklist['(Plaut\. Trin\.|Trin\.)'] = 'phi0119.phi019'
worklist['(Plaut\. Rudens\\b)'] = 'phi0119.phi017'


worklist['(Plin\. N\.\s*H\.)'] = 'phi0978.phi001'


worklist['(Plut[arch]*. adv. Stoic\.|adv. Stoic\.)'] = 'tlg0007.tlg138'
worklist['(Plut[arch]*. Amator\.\s+narr\.|Amator\.\s+narr\.)'] = 'tlg0007.tlg114'
worklist['(Plut[arch]*. Apophth. Lacon\.|Apophth. Lacon\.)'] = 'tlg0007.tlg082'
worklist['(Plut[arch]*. de cohib\. ira\\b|de cohib\. ira\\b)'] = 'tlg0007.tlg095'
worklist['(Plut[arch]*. adv. Colot\.|adv. Colot\.)'] = 'tlg0007.tlg140'
worklist['(Plut[arch]*. de audit\.|de audit\.)'] = 'tlg0007.tlg069'
worklist['(Plut[arch]*. de curios\.|de curios\.)'] = 'tlg0007.tlg102'
worklist['(Plut[arch]*. de educ\.|de educ\.)'] = 'tlg0007.tlg067'
worklist['(Plut[arch]*. de Isid. et Osir\.|de Isid. et Osir\.)'] = 'tlg0007.tlg089'


worklist['(Plut[arch]*. Reip. gerend. praec\.|Reip. gerend. praec\.)'] = 'tlg0007.tlg118'
worklist['(Plut[arch]*. De san. praecept\.)'] = 'tlg0007.tlg077'
worklist['(Plut[arch]*. Defect. [Oo]rac\.|Defect. Orac\.)'] = 'tlg0007.tlg092'
worklist['(Plut[arch]*. de rect\. rat\.|rect\. rat\.)'] = 'tlg0007.tlg069'
worklist['(Plut[arch]*. de ser. num. vind\.|de ser. num. vind\.)'] = 'tlg0007.tlg107'
worklist['(Plut[arch]*. de exilio|de exilio)'] = 'tlg0007.tlg110'
worklist['(Plut[arch]*. de Herodoti malign|de Herodoti malign)'] = 'tlg0007.tlg123'
worklist['(Plut[arch]*. de vitios. pud\.|de vitios. pud\.)'] = 'tlg0007.tlg104'
worklist['(Plut[arch]*. Gryll\.)'] = 'Plutarch-Gryllus'
worklist['((Pseudo-Plut|Plut)[arch]*. X [oO]rat. vit\.|X [oO]rat. vit\.)'] = 'tlg0007.tlg121'
worklist['(Plut[arch]*. Quaest. [Gg]raec\.|Quaest. [Gg]raec\.)'] = 'tlg0007.tlg084b'
worklist['(Plut[arch]*. Symp\.)'] = 'tlg0007.tlg079'




worklist['(Plut[arch]*. Alcib\.)'] = 'tlg0007.tlg015'
worklist['(Plut[arch]*. Arist\.)'] = 'tlg0007.tlg024'
worklist['(Plut[arch]*. Artax\.)'] = 'tlg0007.tlg064'
worklist['(Plut[arch]*. Apophth\.\s+reg\.)'] = 'tlg0007.tlg081'
worklist['(Plut[arch]*.\s+Cat\.\s+Min\.)'] = 'tlg0007.tlg025'
worklist['(Plut[arch]*.\s+Cim\.|Cim\.)'] = 'tlg0007.tlg035'
worklist['(Plut[arch]*.\s+Demetr\.)'] = 'tlg0007.tlg057'
worklist['(Plut[arch]*.\s+Demosth[en]*\.)'] = 'tlg0007.tlg054'
worklist['(Plut[arch]*.\s+Lyc[urg]*\.)'] = 'tlg0007.tlg004'
worklist['(Plut[arch]*.\s+Lys[ander]*\.)'] = 'tlg0007.tlg032'
worklist['(Plut[arch]*.\s+Pelop\.)'] = 'tlg0007.021'
worklist['(Plut[arch]*.\s+Pericl\.)'] = 'tlg0007.tlg012'
worklist['(Plut[arch]*.\s+Sol\.)'] = 'tlg0007.tlg007'
worklist['(Plut[arch]*.\s+Themist\.)'] = 'tlg0007.tlg010'
worklist['(Plut[arch]*.\s+Timol\.)'] = 'tlg0007.tlg018'



worklist['(Poll\.)'] = 'tlg0542.tlg001'

worklist['(Polyb\.|Polybius)'] = 'tlg0543.tlg001'
worklist['(Quintil\.|Quintilian)'] = 'phi1002.phi001'


worklist['(Senec. de const\.)'] = 'stoa0255.stoa009'

 

worklist['(Stob\. Serm\.)'] = 'tlg2037.tlg001'




worklist['(Strab[o]*\.|Strabo\\b)'] = 'tlg0099.tlg001'
worklist['(Suid\.)'] = 'tlg9010.tlg001'

worklist['(Sueton\. Vitell\.)'] = 'phi1348.abo019'

worklist['(Terent.\ Eun[uch]*\.|Eun[uch]*\.)'] = 'phi0134.phi003'
worklist['(Terent.\ Heaut\.|Heaut\.)'] = 'phi0134.phi002'
worklist['(Terent.\ Hecyr\.|Hecyr\.)'] = 'phi0134.phi005'


worklist['(Theocr\.\s+Idyll\.|Theocr\.)'] = 'tlg1714.tlg001'

worklist['((Theophrast\.|Theophr\.)\s+(Char\.|Charact.))'] = 'tlg0093.tlg009'
worklist['((Theophrast\.|Theophr\.)\s+(de caus. plant\.))'] = 'tlg0093.tlg002'
worklist['(Themist\. Or\.)'] = 'tlg2001'
worklist['(Thucyd\.)'] = 'tlg0003.tlg001'

worklist['(Valer. Max\.)'] = 'phi1035.phi001'
worklist['(Vitruv\.)'] = 'phi1056.phi001'



worklist['(Xenoph[ons’\.]+\s+Anab\.)'] = 'tlg0032.tlg006'
worklist['(Xenoph[ons’\.]+\s+Cyropaed\.)'] = 'tlg0032.tlg001'
worklist['(Xenoph[ons’\.]+\s+Hell\.)'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+Oec\.)'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+Memor\.\s+Socr\.|Xenoph\.\s+(Memor\.|Mem\.))'] = 'tlg0032.tlg003'
worklist['(Xenoph[ons’\.]+\s+(de republ\.|de rep. Lac\.))'] = 'tlg0032.tlg010'
worklist['(Xenoph[ons’\.]+\s+Symp\.)'] = 'tlg0032.tlg004'
#worklist['(Xenoph\.\s+)'] = 'tlg0032.tlg003'

def dofile(fname):
    global worklist
    sawauths = 0
    
    romarab = '([XILVC]+,\s+)([0-9\-]+,\s+[0-9\-]+|[0-9\-]+)'
    bookpage = '([XILVC]+,\s+p\.\s+[0-9\-]+)'
    pageref = '\\b(p\.\s+[0-9\-]+)'
    tmpfname = 'chartmp.xml'
    tmpf = open(tmpfname,'w')
    
    f = open(fname)
    intext = 0
    
    text = f.read()
    text = re.sub('\n+','\n',text)
    for l in text.split('\n'):
        l = re.sub('\s+$','',l)
        if(re.search('<text',l)):
            intext = 1
        if(not intext):
            print(l,file=tmpf)
            continue
            
        l = re.sub('^\s+','',l)
        l = re.sub('\s\s+',' ',l)
        l = re.sub('^\s*(<forei[^>]+>)\s*','\g<1>',l)
        l = re.sub('\s*(</foreign>)\s*$','\g<1>',l)
        l = re.sub('(<p>|<div|<p |<cit|<quote|<bibl|<l)','\n\n\g<1>',l)
        l = re.sub('(<p>|<div|<p |<cit|<quote|<bibl)','\n\g<1>',l)
        l = re.sub('(</cit>|</quote>)\s*','\g<1>\n',l)
        
        l = re.sub(bookpage,'<ref n="bookpage">\g<1></ref>',l)
        l = re.sub(romarab,'<ref n="romarab">\g<1>\g<2></ref>',l)
        l = re.sub(pageref,'<ref n="pageref">\g<1></ref>',l)
                
            
        print(l,file=tmpf,end=' ')
    f.close()  
    tmpf.close()

    outfname = re.sub('\.xml','-out.xml',fname)
    print(outfname)
    outf = open(outfname,'w')
    f = open(tmpfname)
    for l in f:
        l = re.sub('\s+$','',l)
        l = re.sub('</foreign>([^<a-zA-Z0-9]+)<foreign xml:lang="grc">','\g<1>',l)
        l = re.sub('([\s,\.;:]+[a-zA-Z][\s,\.;:a-zA-Z]+)(</foreign>)','\g<2>\g<1>',l)
        for foo in worklist:
            m = re.search(foo,l)
            if(m):
                l = re.sub(foo,'<bibl n="'+worklist[foo]+'">\g<1></bibl>',l)
                sawauths = sawauths + 1
        l = re.sub('(<bibl [^>]+>[^<]+)<bibl [^>]+>([^<]+)</bibl>','\g<1>\g<2>',l)
        l = re.sub('(<foreign xml:lang="grc">[^<A-Z:a-z]+)(\s+[:;\.\(a-z\[A-Z,0-9\s]+)','$1</foreign>$2<foreign xml:lang="grc">',l)
        print(l,file=outf)
    outf.close()
    print(sawauths)
    
#dofile('becker-notes.xml')
dofile('becker-appendices.xml')
```
